```python
import pandas as pd
```

```python
one = [0.4, 0.3, 0.9, 0.5, 0.8]
two = [0.5, 0.8, 0.2, 0.6, 0.7]
three = [0.6, 0.9, 0.3, 0.8, 0.5]
```

```python
def multi(one, two):
    result = [[0.0 for x in range(len(one))] for y in range(len(two))]
    for i in range(len(one)):
        for j in range(len(two)):
            result[i][j] = min(one[i],two[j])
            print(str(one[i])+' x '+str(two[j])+' = '+str(result[i][j]))
    return pd.DataFrame(result)
```

```python
def convolute(one, two):
    result = [[0.0 for x in range(len(one))] for y in range(len(two))]
    for x in range(len(one)):
        for y in range(len(one)):        
            result[x][y] = max([min(one[x][y], two[y][x]) for j in range(len(two))])
    return pd.DataFrame(result)
```

```python
matrix1 = multi(one, two)
print(matrix1)
matrix2 = multi(one, three)
print(matrix2)
matrix3 = multi(two, three)
print(matrix3)
matrix1.plot()
```

```python
convo1 = convolute(matrix1, matrix2)
print(convo1)
convo1.plot()
```
